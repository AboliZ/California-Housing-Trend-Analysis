**Features Overview**

---







1. **Longitude:** The horizontal distance of a house measured from west to east. A higher value indicates a more westerly location. It's a continuous quantitative variable.

2. **Latitude:** The vertical distance of a house measured from north to south. A higher value indicates a more northerly location. It's a continuous quantitative variable.

3. **Housing Median Age:** The median age of houses in a block. It's a continuous quantitative variable.

4. **Total Rooms:** The total number of rooms in a block. It's a discrete quantitative variable.

5. **Total Bedrooms:** The total number of bedrooms in a block. It's a discrete quantitative variable.

6. **Population:** The total number of people residing in a block. It's a discrete quantitative variable.

7. **Households:** The total number of households, representing groups of people residing within a home. It's a discrete quantitative variable.

8. **Median Income:** The median income of a household in US Dollars. It's a continuous quantitative variable.

9. **Median House Value:** The median value of households in US Dollars. It's a continuous quantitative variable.

10. **Ocean Proximity:** Indicates the location of the house with respect to the ocean. It's a categorical nominal variable.

**10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.**



```python
# Define a function to categorize the total_bedrooms

def tot_BS(n):
    if n<=10:
       return'small'
    elif n>=11 and n<1000:
       return'medium'
    else:
      return'large'

#Apply the function to create the new column
df["total_bedroom_size"]= df["total_bedrooms"].apply(tot_BS)

# The 'apply' function is used to apply "tot_BS" function to each row in the 'total_bedrooms' column

df
```

```python
df["total_bedroom_size"]
```

```python
# Second Approach

# Define the conditions
conditions = [
    (df['total_bedrooms'] <= 10),
    (df['total_bedrooms'] >= 11) & (df['total_bedrooms'] < 1000),
    (df['total_bedrooms'] >= 1000)]

# Define the categories
categories = ['small', 'medium', 'large']

# Create the new column using numpy
df["total_bedroom_size"] = np.select(conditions, categories)

df
```

 A new column named **'total_bedroom_size'** has been created. It categorizes the total bedrooms as **'small'** if there are 10 or fewer, **'medium'** if there are 11 or more but less than 1000, and **'large'** for totals greater than or equal to 1000.
